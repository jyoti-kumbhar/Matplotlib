```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('data2.csv')
ages = data['Age']
ages.head()
```

```python
bins = [0,10,20,30,40,50,60,70] ##bins are the no. of bars in range u want in chart 
plt.hist(ages, bins = bins, edgecolor = 'black')
plt.title('Histogram Chart')
plt.xlabel('Ages')
plt.ylabel('Total respondents')
#plt.grid()
plt.tight_layout()
```

```python
bins = [0,10,20,30,40,50,60,70] ##bins are the no. of bars in range u want in chart 
plt.hist(ages, bins = bins, edgecolor = 'black', log=True)
plt.title('Histogram Chart')
plt.xlabel('Ages')
plt.ylabel('Total respondents')
#plt.grid()
plt.tight_layout()
```

```python
bins = [0,10,20,30,40,50,60,70] ##bins are the no. of bars in range u want in chart 
plt.hist(ages, bins = bins, edgecolor = 'black')
median_age = 29
plt.axvline(median_age, color='red', label='Median Age', linewidth = 2)
plt.title('Histogram Chart')
plt.xlabel('Ages')
plt.ylabel('Total respondents')
#plt.grid()
plt.legend()
plt.tight_layout()
```

```python
#demo7

x=[1, 2, 3, 4, 5, 6, 7, 8, 9, 2, 4, 6, 8, 1, 3]
y=[9, 8, 7, 6, 5, 4, 3, 2, 1, 9, 7, 5, 3, 8, 6]
# s is the size
plt.scatter(x,y,s=100, c='red', linewidth=1, alpha=0.75, edgecolors='black')
plt.title('Scatter Chart')
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()
```

```python
x=[1, 2, 3, 4, 5, 6, 7, 8, 9, 2, 4, 6, 8, 1, 3]
y=[9, 8, 7, 6, 5, 4, 3, 2, 1, 9, 7, 5, 3, 8, 6]
colors = [3, 1, 7, 6, 4, 2, 8, 9, 5, 7, 1, 3, 6, 9, 4]

plt.scatter(x,y,s=100, c=colors, cmap='Greens', linewidth=1, alpha=0.75, edgecolors='black')
# adds a color bar
cbar = plt.colorbar()
cbar.set_label('Satisfication')
plt.title('Scatter Chart')
plt.xlabel('x')
plt.ylabel('y')
plt.tight_layout()
```

```python
data = pd.read_csv('data3.csv')
data.head()
```

```python
likes = data['Likes']
views = data['Views']
ratio = data['Ratio']
```

```python
plt.scatter(views, likes, c=ratio, cmap='Oranges', linewidth=1, alpha=0.75, edgecolors='black')
cbar = plt.colorbar()
cbar.set_label('Ratio')
plt.title('Trending videos')
plt.xlabel('View Count')
plt.ylabel('Total Likes')
plt.tight_layout()
```
