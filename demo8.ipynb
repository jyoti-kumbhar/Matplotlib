```python
import pandas as pd
from matplotlib import pyplot as plt
from matplotlib import dates as mpl_dates
from datetime import datetime, timedelta
```

```python
dates = [
    datetime(2025, 4, 5),
    datetime(2025, 4, 12),
    datetime(2025, 4, 17),
    datetime(2025, 4, 23),
    datetime(2025, 4, 28),
    datetime(2025, 4, 20),
    datetime(2025, 4, 30)
]
y=[3, 7, 0, 5, 1, 6, 4]
```

```python
plt.plot_date(dates, y, linestyle='solid')
plt.tight_layout()
```

```python
plt.plot_date(dates, y, linestyle='solid')
plt.tight_layout()
# gcf = give current figure this rotates the  x axis's labels
plt.gcf().autofmt_xdate()
```

```python
plt.plot_date(dates, y, linestyle='solid')
plt.tight_layout()
plt.gcf().autofmt_xdate()
dateFormate = mpl_dates.DateFormatter('%b, %d, %Y')
#gca = get current access
plt.gca().xaxis.set_major_formatter(dateFormate)
```

```python
data = pd.read_csv('data4.csv')
Data = data.head(20)
```

```python
Data.head()
```

```python
Data['Date'] = pd.to_datetime(Data['Date'])
Data.sort_values('Date', inplace = True)
date=Data['Date']
close=Data['Close']
```

```python
plt.plot_date(date, close, linestyle='solid')
plt.title('Time plot')
plt.xlabel('Date')
plt.ylabel('Close Price')
plt.tight_layout()
plt.gcf().autofmt_xdate()
dateFormate = mpl_dates.DateFormatter('%b, %d, %Y')
plt.gca().xaxis.set_major_formatter(dateFormate)
```
