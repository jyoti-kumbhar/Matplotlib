```python
from matplotlib import pyplot as plt
import numpy as np
import csv
```

```python
ages = [25, 30, 35, 40, 28, 32, 45, 29, 50, 26]
# Sorted salary lists
pysalaries = [50000, 52000, 55000, 58000, 60000, 65000, 70000, 80000, 90000, 100000]
jssalaries = [49000, 54000, 57000, 59000, 62000, 67000, 72000, 82000, 92000, 102000]
javasalaries = [53000, 55000, 58000, 60000, 63000, 68000, 73000, 83000, 93000, 103000]
```

```python
#bar chart
plt.bar(ages, pysalaries,color='#444444', label='python developers')

plt.title('Plot')
plt.xlabel('Age')
plt.ylabel('Salary')
```

```python
plt.bar(ages, pysalaries,color='#444444', label='python developers')
plt.plot(ages, jssalaries,color='m', label='js developers')

plt.title('Plot')
plt.xlabel('Age')
plt.ylabel('Salary')
```

```python
x_indexes = np.arange(len(ages))
```

```python
width=0.30
plt.bar(x_indexes - width, pysalaries, width = width, color='#444444')
plt.bar(x_indexes, jssalaries,width = width, color='m')
plt.bar(x_indexes + width, javasalaries,width = width, color='b')

plt.xticks(ticks=x_indexes, labels=ages)
plt.title('Plot')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend(['python developers','js developers','java developers'])
```

```python
with open('data1.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    #row = next(csv_reader)
    #print(row['LanguagesWorkedWith'].split(';'))

    language_counter = Counter()
    for row in csv_reader:
        language_counter.update(row['LanguagesWorkedWith'].split(';'))

languages = []
popularity = []
for i in language_counter.most_common(15):
    languages.append(i[0])
    popularity.append(i[1])
print(languages, popularity)
```

```python
#horizontal bar chart
plt.barh(languages,popularity)
plt.title('Popular Language')
#plt.ylabel('Language')
plt.xlabel('Popularity')
plt.tight_layout()
plt.grid()
```

```python
languages.reverse()
popularity.reverse()
plt.barh(languages,popularity)
plt.title('Popular Language')
#plt.ylabel('Language')
plt.xlabel('Popularity')
plt.tight_layout()
plt.grid()
```
