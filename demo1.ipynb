```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
data=df.head()
data
```

```python
data.shape
```

```python
xaxis = data['High']
xaxis
```

```python
xaxis1 = data['Low']
xaxis1
```

```python
yaxis = data['Open']
yaxis
```

```python
plt.plot(xaxis, yaxis,color='#444444', linestyle='--', marker='.', linewidth=2)##k is the black color and -- is the dotted lines
plt.plot(xaxis1, yaxis,'.m-.',linewidth=2)

plt.title('Plot')
plt.xlabel('High and Low')
plt.ylabel('Open')
plt.legend(['High','Low'])
plt.grid()
plt.tight_layout()#adjusts the padding
```

```python
plt.plot(xaxis, yaxis)
plt.plot(xaxis1, yaxis)

plt.style.use('ggplot')
plt.title('Plot')
plt.xlabel('High and Low')
plt.ylabel('Open')
plt.legend(['High','Low'])
plt.savefig('plt.png')##to save the image give path with file name
```
